```python
'''
7. Створіть дві матриці розміром 2x2 та 2x3, заповніть їх випадковими 
цілими числами в діапазоні від 1 до 10 та перемножте їх між собою.
'''

matrix_1 = np.random.randint(1, 11, (2, 2))
matrix_2 = np.random.randint(1, 11, (2, 3)) 

print("Матриця 1: \n", matrix_1)
print("\n Матриця 2: \n", matrix_2)
print("\n Матриця 1 * Матриця 2: \n", np.matmul(matrix_1, matrix_2))
```

```python
'''
8. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами 
в діапазоні від 1 до 10 та знайдіть її обернену матрицю.
'''

matrix_a = np.random.randint(1, 11, (3, 3))
matrix_b = matrix_a.T

print(f'Матриця а: \n{matrix_a}')
print(f'\n Матриця, обернена до неї: \n{matrix_b}')  
```

```python
'''
16. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні 
від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
'''

a = np.random.rand(3, 3)
vector = a.sum(axis=1).reshape(-1, 1)

print('Матриця:')
print(a)
print('\nВектор:')
print(vector) 
```

```python
from surprise import SVD, SVDpp, NMF
from surprise import Dataset
from surprise.model_selection import cross_validate, GridSearchCV

data = Dataset.load_builtin('ml-100k')

# Use the famous SVD algorithm.
algo = SVD()

# Run 5-fold cross-validation and print results.
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
# Підбір найкращих параметрів для SVD за допомогою GridSearchCV
print("\nПошук найкращих параметрів для SVD:")
param_grid = {'n_factors': [20, 50, 100], 'lr_all': [0.002, 0.005, 0.01], 'reg_all': [0.02, 0.1, 0.2]}
gs_svd = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)
gs_svd.fit(data)

print("Найкраще значення RMSE:", gs_svd.best_score['rmse'])
print("Найкращі параметри для SVD:", gs_svd.best_params['rmse'])
```

```python
# алгоритм SVD++
print("\nРезультати для SVD++:")
algo_svdpp = SVDpp()
cross_validate(algo_svdpp, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
# алгоритм NMF
print("\nРезультати для NMF:")
algo_nmf = NMF()
cross_validate(algo_nmf, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)


```

```python
print("\nПошук найкращих параметрів для NMF:")
param_grid_nmf = {'n_factors': [20, 50, 100], 'lr_bu': [0.002, 0.005, 0.01], 'reg_bu': [0.02, 0.1, 0.2]}
gs_nmf = GridSearchCV(NMF, param_grid_nmf, measures=['rmse', 'mae'], cv=3)
gs_nmf.fit(data)

print("Найкраще значення RMSE:", gs_nmf.best_score['rmse'])
print("Найкращі параметри для NMF:", gs_nmf.best_params['rmse'])
```

Оптимальним алгоритмом є SVD++, але його швидкість дуже повільна.

Середнє значення RMSE: 0.9185 (там де std в результатах комірки 4)

Середнє значення MAE: 0.7214

Для завдань де треба швидший час навчання краще використовувати SVD
